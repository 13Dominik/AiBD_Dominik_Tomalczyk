```python
# Zmienione nazwy kolumn
columns = ['Worried_about_earthq', "How_worried_Babout_Big_One",'Big_One_will_occur', 
           'earthq_experienced', 'precautions', 'familiar_with_SA_Fault_time',
           'familiar_with_yellowstone_supervolcano',
           "Age", "Sex", "Household_year_earn", "US_region"]
df = pd.read_csv('../Analysis_Data/earthquake_data.csv', names=columns, header=0)
```

```python
df
```

```python
# Tabela łącząca wiek, płeć i odpowiedź na pytanie "Do you think the "Big One" will occur in your lifetime?"
df = df[["Age", "Sex", "Big_One_will_occur"]]
```

```python
df.groupby("Sex").count()
```

```python
# Zapis danych do folderu z danymi.
df.to_csv('../Analysis_Data/earthquake_data.csv')
```

 ## Visualization

```python
sns.set_style('darkgrid')
sns.countplot(df, x="Big_One_will_occur")
plt.title("""Will "Big One" Occur in your lifetime?""")
```

```python
sns.countplot(df, x="Big_One_will_occur", hue="Sex")
plt.title("""Will "Big One" Occur in your lifetime?""")
```
