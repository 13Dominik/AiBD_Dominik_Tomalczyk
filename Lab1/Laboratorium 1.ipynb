```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
f = lambda x: x**2 + 5
```

```python
x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-6, 6, 100)
x3 = np.linspace(0, 5, 100)
```

```python
plt.plot(x1, f(x1), label = "x^2 + 5")
plt.title("Wykres funkcji x^2 + 5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
plt.plot(x2, f(x2), label = "x^2 + 5")
plt.title("Wykres funkcji x^2 + 5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
plt.plot(x3, f(x3), label = "x^2 + 5")
plt.title("Wykres funkcji x^2 + 5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
